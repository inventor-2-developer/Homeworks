```python
import pandas as pd


# 1. Printing the bar

team_1=pd.read_csv('team-1.csv')
team_1_data=pd.DataFrame(team_1)
team_1_data.plot.bar()
```

```python
# 2 Melting the DataFrame

pd.melt(team_1_data)

#pd.melt(team_1_data,id_vars=['runner_name'],value_vars=['run_1','run_2','run_3','run_4'],var_name='coach_name')
```

```python
# 3 Melting the DataFrame using the "runner_name"column
pd.melt(team_1_data,id_vars=['runner_name'])
```

```python
# 4 Melting the DataFrame using the "coach_name"column
pd.melt(team_1_data,id_vars=['coach_name'])
```

```python
#5 Removed the "coach_name" from the original dataframe
new_frame=team_1_data.drop(columns=['coach_name'])
new_frame
```

```python
#6 Melting the updated by using the "runner_name" column and then add a var_name and a value_name

updated_frame=pd.melt(new_frame,id_vars=['runner_name'],var_name='run',value_name='distance')
updated_frame
```

```python
#7 Create another bar chart using the melt function
updated_frame.plot.bar()
```

```python
#8 Save the melted DataFrame to a CSV file
updated_frame.to_csv('Homework_4_csv')
```

```python
#9 Reorient the melted DataFrame to match the original CSV
updated_frame.pivot_table(values='distance',index='runner_name',columns='run')
```
