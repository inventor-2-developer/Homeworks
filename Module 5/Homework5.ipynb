```python
import pandas as pd
import os

# Removed the warnings because I kept getting them
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# 1 Import the CSV file into a Dataframe, and then get a statistical summary of all numeric values.
data_file = os.path.join('pools.csv')
data_file_df = pd.read_csv(data_file)
data_file_df.describe()
```

```python
# 2 Retrieve a brief summary of the DataFrame.
data_file_df.info()
```

```python
# 3 Return the first 7 rows in the DataFrame.
data_file_df.head(7)
```

```python
#4 Return the last 10 rows in the DataFrame.
data_file_df.tail(10)
```

```python
#5 Find the total number of boroughs 
data_file_df["BOROUGH"].value_counts
```

```python
#print the unique borough codes.
unique=data_file_df["BOROUGH"].unique()
unique
```

```python
#6 The total area of all the pools is retrieved.
total_area=data_file_df["SHAPE_STAr"].sum()
total_area
```

```python
#7 The average area of all the pools is retrieved.
average_area=data_file_df["SHAPE_STAr"]
average_area.mean()
```

```python
# 8 Find the total number of pools in each borough.
xcounter=0
qcounter=0
bcounter=0
mcounter=0
rcounter=0
for x in data_file_df["BOROUGH"]:
    if x=="X":
        xcounter=xcounter+1
    if x=="Q":
        qcounter=qcounter+1
    if x=="B":
        bcounter=bcounter+1
    if x=="M":
        mcounter=mcounter+1
    if x=="R":
        rcounter=rcounter+1
print(f"There are {xcounter} pools in Borough X,"f" {qcounter} pools in Borough Q, "f" {bcounter} pools in Borough B,"
      f" {mcounter} pools in Borough M," f" and {rcounter} pools in Borough R."f" This totals to {bcounter+qcounter+mcounter+rcounter+xcounter} pools.")
      
   
```

```python
#9 The total area of all the pools in each borough is retrieved by using groupby.
data_file_df.groupby('BOROUGH')['OBJECTID','COUNCILDIS','SHAPE_STAr','SHAPE_STLe'].agg("sum")
```

```python
#10 The average area of all the pools in each borough is retrieved by using groupby.
data_file_df.groupby('BOROUGH')['OBJECTID','COUNCILDIS','SHAPE_STAr','SHAPE_STLe'].agg("mean")
```

```python
# Bonus: A pivot table is created to retrieve the total area of all the different pool types for each borough.

df=pd.pivot_table(data_file_df,index='BOROUGH',columns=['POOLTYPE'],aggfunc={'SHAPE_STAr':"sum"})
df
```
