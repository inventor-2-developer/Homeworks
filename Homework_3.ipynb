```python
#importing pandas
import pandas as pd
data=pd.read_csv('ratings_and_sentiments.csv')
#Dropping columns not needed
dz=data.drop(columns=['vibe_sent','rating','cat_rating','bool_HIGH','tea_sent','service_sent','price_sent','parking_sent','food_sent','local_sent','internet_sent','hours_sent','alcohol_sent','seating_sent','location_sent','coffee_sent'])
#Creating the Dataframe
df=pd.DataFrame(dz)
df 
```

```python
#Returning all of the shops that have number rating greater than or equal to 4
df[df['num_rating']>=4]
```

```python
#Returning all of the shops that have an overall sentiment greater than or equal to 4
df[df['overall_sent']>=4]
```

```python
#Finding the number of the shops that have an overall sentiment greater than or equal to 4 and number rating greater than 4
dp=df[(df['overall_sent']>=4)&(df['num_rating']>4)]
print("There are " + f"{dp['coffee_shop_name'].count()}" " " +"coffee shops in this list.")
```

```python
# Changing the names of the review text column to comments and the overall sent to sentiment rating
rename_df=dp.rename(columns={'review_text':'comments','overall_sent':'sentiment_rating'})
rename_df
```

```python
df.to_csv('new_file.csv', sep='\t', index=False)
```
